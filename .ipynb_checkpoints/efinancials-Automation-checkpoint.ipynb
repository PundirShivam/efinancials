```python
def get_web_links(browser,class_name=None):
    if class_name is None:
        class_name = 'jobs-unified-top-card__content--two-pane'
    web_links = []
    for ele in browser.find_element(By.CLASS_NAME,class_name).find_elements(By.CSS_SELECTOR,'a'):
        web_links.append(ele.get_attribute('href'))
    return web_links

def get_company_link(browser):
    try:
        html = browser.find_element(By.CLASS_NAME,'job-details-jobs-unified-top-card__primary-description-container').get_attribute('innerHTML')
        soup = BeautifulSoup(html, 'html.parser')
        link = soup.find('a').get('href')    
        return link
    except:
        return
    
def get_job_link(element):
    try:
        soup = BeautifulSoup(element.get_attribute('innerHTML'),'html.parser')
        link = 'www.linkedin.com'+soup.find('a').get('href')
        link = link.split('/?')[0]
        return link
    except e:
        print('job link cannot be extracted')
        
def get_job_description(browser,job_description_class=None):
    """"""
    if job_description_class is None: 
        job_description_class = 'jobs-description__container'
    try:
        job_text = browser.find_element(By.CLASS_NAME,job_description_class).text
#        job_text = job_text.split('\n')
        return job_text
    except:
        return None

def click_job_and_get_data(browers,xpath=None,sleep_time=2):
    
    if xpath is None:
        xpath = "//div[@data-job-id]"
    
    
    elements = browser.find_elements(By.XPATH,xpath)
    
    info_list = []
    for ele in elements:
        ele.click()
        time.sleep(sleep_time)
        
        try:
            job_description = get_job_description(browser)

            company_name    = get_job_description(browser,'job-details-jobs-unified-top-card__primary-description-container')

            job_title_name  = get_job_title(browser)

            company_link    = get_company_link(browser)

            job_link        = get_job_link(ele)

            hirer_info , hirer_profile_link = get_hirer_info(browser)

            info_list.append([job_title_name,company_name,job_description,job_link,company_link,hirer_info , hirer_profile_link])
            
            print(f'{job_title_name} @ {company_name}')
            print('--'*50)
    
        except  e:
            print(e)
            continue

    return info_list

def get_max_page_count(browser,class_name=None):
    if class_name is None:
        class_name = 'artdeco-pagination__pages.artdeco-pagination__pages--number'
    max_page_number = browser.find_element(By.CLASS_NAME,class_name).text.split('\n')
    max_page_number = pd.to_numeric(pd.Series(max_page_number),errors='coerce').max()
    return int(max_page_number)

def scroll_down_all_jobs_on_page(browser):
    class_master_uk_list = 'scaffold-layout__list-container'
    elements = browser.find_element(By.CLASS_NAME,class_master_uk_list).find_elements(By.TAG_NAME,'li')
    for i in range(1,len(elements)):
        try:
            ActionChains(browser).move_to_element(elements[i]).perform()
        except:
            time.sleep(3)
            print('moving page downwards')
            
        
def get_job_title(brower,class_name=None):
    if class_name is None:
        class_name = 'job-details-jobs-unified-top-card__job-title-link'
    return browser.find_element(By.CLASS_NAME,class_name).text

def get_hirer_info(browser):

    elements = browser.find_elements(By.CLASS_NAME,'hirer-card__container')
    if len(elements)==0:
        return None,None
    soup = BeautifulSoup(elements[0].get_attribute('innerHTML'),'html.parser')
    return elements[0].text,soup.find('a').get('href')
```

```python
def click_the_button(browser,page_number):
    try:
        browser.find_element(By.XPATH,"//li[@data-test-pagination-page-btn='{}']//button".format(page_number)).click()
    except:
        browser.find_element(By.XPATH,'//button[@aria-label="Page {}"]'.format(page_number)).click()
```

```python

#time.sleep(1)
max_pages = get_max_page_count(browser)
current_page = 1
data_list = []
```

```python
print('Max no of pages {}'.format(max_pages))
while current_page <= max_pages :
    print('Working on Page # {} '.format(current_page))    
    scroll_down_all_jobs_on_page(browser)
    try:
        tmp_data = click_job_and_get_data(browser)
    except:
        time.sleep(3)
        tmp_data = click_job_and_get_data(browser)        
    data_list.append(tmp_data)
    current_page = current_page + 1
    if current_page == max_pages:
        break
    click_the_button(browser,page_number=current_page)
    
```

```python
data = pd.concat((pd.DataFrame(data) for data in data_list),axis=0)

data = data.reset_index(drop=True)

columns = ['JobTitle','Company','JobDescription','JobLink','CompanyTitle','JobPoster','JobPosterProfile']
data.columns = columns

data = data.drop_duplicates().reset_index(drop=True)
```
